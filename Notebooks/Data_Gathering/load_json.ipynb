```python
import pandas as pd

dataset_id = "lkytal/trainjson"
file_path = "train.json"

# Step 1: Download dataset locally (returns path)
local_path = kagglehub.dataset_download(dataset_id)

# Step 2: Build full file path (inside the dataset folder)
full_path = f"{local_path}/{file_path}"

# Step 3: Read JSON safely (handle encoding issues)
df = pd.read_json(full_path, encoding='latin-1')  # or try 'ISO-8859-1'
df.head()
```

```python
import json
import pandas as pd

# full_path = "train.json"  # your file path

# Step 1: Load JSON
with open(full_path, 'r', encoding='latin-1') as f:
    data = json.load(f)  # loads as a dict or list

# Step 2: Normalize into a flat DataFrame
df = pd.json_normalize(data)

# Step 3: Check columns
df.head()
df.columns
```

```python
band1_df = pd.DataFrame(df['band_1'].to_list())
band2_df = pd.DataFrame(df['band_2'].to_list())

# Concatenate with other columns
full_df = pd.concat([df[['id','inc_angle','is_iceberg']], band1_df, band2_df], axis=1)
df.head()
```
