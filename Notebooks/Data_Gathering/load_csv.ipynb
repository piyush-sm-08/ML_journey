Converters :


```python
def rename(name):
    if not isinstance(name, str) or not name.strip():
        return name  
    name = name.strip()
    initial = name[0].upper()
    for i, ch in enumerate(name):
        if ch == ' ' and i+1 < len(name):
            initial += name[i+1].upper()
    return initial
```

```python
pd.read_csv('csv_file/matches.csv')['team1'].apply(rename)
```

         method-1

```python
pd.read_csv('csv_file/matches.csv',converters={'team1':rename,
                                               'team2':rename,
                                               'toss_winner':rename,
                                               'winner':rename})
```

         method-2

```python
pd.read_csv('csv_file/matches.csv').assign(
    team1=lambda x: x['team1'].apply(rename),
    team2=lambda x: x['team2'].apply(rename),
    toss_winner=lambda x: x['toss_winner'].apply(rename),
    winner=lambda x : x['winner'].apply(rename)
)
```
