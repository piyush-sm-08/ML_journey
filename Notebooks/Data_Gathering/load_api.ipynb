```python
import requests
import time
import requests
import pandas as pd
import time

session = requests.Session()
session.headers.update(headers)

df_list = []
columns = ['id','original_title','overview','release_date','original_language','popularity','vote_count','vote_average']

for i in range(1, 429):
    for attempt in range(5):
        try:
            url = f"https://api.themoviedb.org/3/movie/top_rated?language=en-US&page={i}"
            response = session.get(url, timeout=10)
            response.raise_for_status()
            temp_df = pd.DataFrame(response.json()['results'])[columns]
            df_list.append(temp_df)
            time.sleep(0.8)  # delay after successful request
            break
        except requests.exceptions.RequestException as e:
            wait = 2 ** attempt  # exponential backoff
            print(f"Page {i} failed on attempt {attempt+1}: {e}. Retrying in {wait}s...")
            time.sleep(wait)
```

```python
df=pd.concat(df_list,ignore_index=True)
```
